# Definizione finestra di gioco e aggiornamento

```python
import turtle
from sys import platform


wn = turtle.Screen() # crea una finestra nuova (sarà lo schermo del gioco)
wn.title('Pong by @TokyoEdTech') # inserisci titolo della scheda nuova wn
wn.bgcolor('black') # colore delo sfondo di wn
wn.setup(width= 800, height=600) # definisce le dimensioni della finestra
wn.tracer(8) # 'impedisce alla finestra creata di aggiornarsi continuamente', se non settato fa crashare la finestra 

# Score
score_a = 0
score_b = 0

# Paddle A
paddle_a = turtle.Turtle()
paddle_a.speed(0) # velocità di animazione del paddle_a
paddle_a.shape('square') # definiamo qua la dimensione 'square' per il paddle_a
paddle_a.color('white') # colore bianco del baddle_a
paddle_a.shapesize(stretch_wid=5, stretch_len=1) # questo comando va ad allungare lo shape del paddle_a per l'asse y e l'asse x
paddle_a.penup() # 'solleva' il paddle_a dallo sfondo. Altrimenti viene disegnata una linea che lo collega al centro dello schermo
paddle_a.goto(-350, 0) # diamo la posizione di spawn del paddle_a (NB: (0,0) è il centro della finestra che disegnamo)

# Paddle B
paddle_b = turtle.Turtle()
paddle_b.speed(10) 
paddle_b.shape('square') 
paddle_b.color('white') 
paddle_b.shapesize(stretch_wid=5, stretch_len=1) # NB: i valori vanno moltiplicati per 20 nella finestra (solo per questa funzione)
paddle_b.penup()
paddle_b.goto(350, 0) 

# Ball
ball = turtle.Turtle()
ball.speed(0) 
ball.shape('circle') 
ball.color('white') 
ball.penup()
ball.goto(0, 0) 
ball.dx = 0.1 # velocità con cui la palla si muove verso destra (asse x+)
ball.dy = 0.1 # velocità con cui la palla si muove in alto (asse y+)


# Pen: contatore punteggio
pen = turtle.Turtle() # creo un oggetto di classe turtle
pen.speed(0) # velocità animazione
pen.color('white')
pen.penup()
pen.hideturtle() # nascondiamo lo score
pen.goto(0,260)
pen.write('Player A: 0  Player B: 0'   # scrive una riga di testo
            ,align = 'center'          # dove vuoi che appaia il testo
            ,font = ('Courier', 24, 'normal')) # stile, font, grassetto o normale 


# funzioni di gioco
def paddle_a_up():   # definiamo la funzione per muovere verso l'alto il paddle_a
    y = paddle_a.ycor() # prende la coordinata y attuale del paddle_a
    y += 20 # definiamo di quando si muove il paddle_a per ogni volta che vogliamo salire
    paddle_a.sety(y) # aggiorna il valore della coordinata y

def paddle_a_down():   
    y = paddle_a.ycor() 
    y -= 20 
    paddle_a.sety(y) 

def paddle_b_up():   
    y = paddle_b.ycor() 
    y += 20 
    paddle_b.sety(y) 

def paddle_b_down():   
    y = paddle_b.ycor() 
    y -= 20 
    paddle_b.sety(y) 

# Keyboard binding
# sono i comandi della tastiera che consentono di richiamare le funzioni 
wn.listen() # il comando 'listen' dice allo script di leggere i comandi da tastiera   
wn.onkeypress(paddle_a_up, 'w') # se premi il tasto w, richiama la funzione paddle_a_up
wn.onkeypress(paddle_a_down, 's')
wn.onkeypress(paddle_b_up, 'Up') # NB: per il paddle_b usiamo le freccie direzionali, attraverso la stringa speciale 'Up' e 'Down'
wn.onkeypress(paddle_b_down, 'Down')



# Main Game Loop: ciclo di aggiornamento del gioco
game_on = True
while game_on == True:
    wn.update() # aggiorna la finestra continuamente ad ogni loop

    # Movimento palla
    ball.setx(ball.xcor() + ball.dx) # aggiunge alla coordinata X della palla il valore ball.dx, generando il movimento
    ball.sety(ball.ycor() + ball.dy) # stesso per Y

    # Controllo bordi: nè la palla nè i paddle possono uscire dallo schermo!
    if ball.ycor() > 290: # NB: per l'asse y, la dimensione della finestra è (-300, 300), pertanto lasciamo una tolleranza di 10 da ogni bordo
        ball.sety(290)
        ball.dy *= -1 # inverte la direzione Y della balla!

    if ball.ycor() < -290: 
        ball.sety(-290)
        ball.dy *= -1 

    if ball.xcor() > 390: # NB: asse X = (-400, 400), ma se raggiunge il borso, allora è un punto per uno dei due giocatori!
        #ball.goto(0, 0) # riposiziona la palla al centro
        ball.dx *= -1
        score_a += 1
        pen.clear()
        pen.write('Player A: {}  Player B: {}'.format(score_a, score_b)   
            ,align = 'center'          
            ,font = ('Courier', 24, 'normal')) 
 

    if ball.xcor() < -390: 
        #ball.goto(0, 0)
        ball.dx *= -1 
        score_b += 1
        pen.clear() # pulisce l'oggetto pen togliendolo. necessario prima di poterlo sostituire col nuovo score
        pen.write('Player A: {}  Player B: {}'.format(score_a, score_b)   
            ,align = 'center'          
            ,font = ('Courier', 24, 'normal')) 

    # Collisione tra palla e paddle
    # Ogni paddle ha centro -350 o 350, ed è un rettagolo 100*20 (20*5, 20*1) centrato sull'asse 350 o -350.
    # Per definire la collisione con la palla (che è un 20*20 di dimensione), bisogna calcolare le coordinate dei bordi del paddle.
    # NB: conviene fare diversi tentativi con i bordi, per calibrare bene se la collisione è sensata dal punto di vista grafico.
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() + 50 and ball.ycor() > paddle_b.ycor() - 40):
        ball.setx(340)
        ball.dx *= -1 

    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() + 50 and ball.ycor() > paddle_a.ycor() - 40):
        ball.setx(-340)
        ball.dx *= -1 

    if score_a == 10:
        game_on = False
        pen.clear()
        pen.write('Player A vince!'   # scrive una riga di testo
            ,align = 'center'          # dove vuoi che appaia il testo
            ,font = ('Courier', 24, 'normal')) # stile, font, grassetto o normale 
        turtle.exitonclick()
    if score_b == 10:
        game_on = False
        pen.clear()
        pen.write('Player B vince!'   # scrive una riga di testo
            ,align = 'center'          # dove vuoi che appaia il testo
            ,font = ('Courier', 24, 'normal')) # stile, font, grassetto o normale 
        turtle.exitonclick() # al premere del muose chiude la finestra

turtle.mainloop()


```
